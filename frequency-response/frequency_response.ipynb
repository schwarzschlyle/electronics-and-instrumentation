```python
import pandas as pd
import plotly.express as px
from datetime import datetime
from google.colab import files
import io
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.preprocessing import MinMaxScaler
import seaborn as sb
import statistics 


import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(10, 6))



d_test_df = pd.read_csv('d_test.csv')
d_train_df = pd.read_csv('d_train.csv')
d_sa_df = pd.read_csv('d_train_sa.csv')


plt.plot(d_train_df['Frequency (Hz)'], d_train_df['Level (dB)'], label='Train data', c = "red")
plt.plot(d_sa_df['Frequency (Hz)'], d_sa_df['Level (dB)'], label='Train data with stochastic amplification', c = "blue")

plt.axvline(x = 196, color = 'r', linestyle = '-', label = 'G string empirical frequency', c = "black")

plt.xscale('log')

plt.title("Frequency content data of G string", size=12)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Level (dB)")
plt.legend(loc='lower right')
```

# XGBoost

```python
import xgboost as xgb


a_y_test = a_test_df['Level (dB)']
a_x_test = a_test_df[['Frequency (Hz)']]

a_y_train = a_train_df['Level (dB)']
a_x_train = a_train_df[['Frequency (Hz)']]


train_reg = xgb.XGBRegressor().fit(a_x_train, a_y_train)
a_train_score = train_reg.score(a_x_test, a_y_test)
xga_a_train_pred = train_reg.predict(a_x_train)

a_y_sa = a_sa_df['Level (dB)']
a_x_sa = a_sa_df[['Frequency (Hz)']]

sa_reg = xgb.XGBRegressor().fit(a_x_sa, a_y_sa)
a_sa_score = sa_reg.score(a_x_test, a_y_test)
xga_a_sa_pred = sa_reg.predict(a_x_sa)




plt.figure(figsize=(10, 6))
plt.title("XGBoost fit for B string frequency")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Level (dB)")
plt.xscale('log')

plt.scatter(a_test_df[['Frequency (Hz)']], a_test_df['Level (dB)'], c='gray', label="Test data")
plt.plot(a_x_train, xga_a_train_pred, c="red", label="XGB fit using train data")
plt.plot(a_x_sa, xga_a_sa_pred, c="blue", label="XGB fit using train data with stochastic amplification")
plt.plot(a_x_sa, abs(xga_a_sa_pred-xga_a_train_pred), c="black", label="Absolute difference due to schoastic amplification", linestyle = "dashed")
plt.axvline(x = 247, color = 'r', linestyle = '-', c="black", label="B string empirical frequency (247 Hz)")

plt.legend(loc='lower left')
plt.show()



print(a_train_score)
print(a_sa_score)

print(xga_a_sa_pred.argmax())
print(a_x_sa.iloc[xga_a_sa_pred.argmax()])

print(xga_a_train_pred.argmax())
print(a_x_train.iloc[xga_a_train_pred.argmax()])



```
