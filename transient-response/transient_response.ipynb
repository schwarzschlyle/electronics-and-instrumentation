```python
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(12, 8))



plt.plot(hot_df['time_t'], hot_df['data'], label='Heating curve')



plt.title("Temperature time series data of heating step response", size=12)
plt.xlabel("Time (Seconds)")
plt.ylabel("Temperature (Celcius)")
plt.legend(loc='lower right')
```
